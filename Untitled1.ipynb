```python
#prompt : import pandas as pd

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df=pd.read_csv("/content/diabetes (2).csv")
```

```python
x=df.drop(columns=['Outcome'])
y=df['Outcome']
```

```python
import matplotlib.pyplot as plt
plt.hist(df[df['Outcome']==0]['Glucose'],bins=20,alpha=0.5,label="Non-diabetic(0)",color='blue')
plt.hist(df[df['Outcome']==1]['Glucose'],bins=20,alpha=0.5,label="Diabetic(1)",color='red')
plt.xlabel("Glucose Level")
plt.ylabel("Count")
plt.title(" Glucose Level Distribution: Diabetic vs Non-Diabetic Patients")
plt.legend()
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scalar=StandardScaler()
x_train_scaled=scalar.fit_transform(x_train)
x_test_scaled=scalar.transform(x_test)
```

```python
#Train a SVM classifier
model=SVC(kernel='linear',random_state=42)
model.fit(x_train_scaled,y_train)

#Make Predictions
y_pred=model.predict(x_test_scaled)
```

```python
accuracy=accuracy_score(y_test,y_pred)*100
print(f"\nModel accuracy:{accuracy:.2f}%")
```

```python
#Classification report
print("\nClassification Report:\n",classification_report(y_test,y_pred))
```

```python
print("\nConfusion Matrix:\n",confusion_matrix(y_test,y_pred))
```
